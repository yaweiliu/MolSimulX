# Example: control the local pressure in a mechanical way

* In a system with interfaces (i.e., multiphase system), the pressure is not homogeneous and the pressure in the interfaical region normally is lower than that in the bulk region, so we cannot maintin the desired pressure in the bulk region with a global barostat. In this case, a mechanical way can be used to control the pressure. See [J. Chem. Phys. 2018, 148, 064706](https://doi.org/10.1063/1.5011106) for the disscusion of global and local pressure controls in MD simulations.
* This method has been widly used in our previous work such as [J. Chem. Phys. 2014, 141, 134702]( https://doi.org/10.1063/1.4896937) and [J. Phys. Condens. Matter 2018, 30, 205002](https://doi.org/10.1088/1361-648X/aabd58)

```python
%reload_ext autoreload
%autoreload 2
import numpy as np
import pandas as pd
pd.options.display.max_rows = 6
pd.options.display.max_columns = None
pd.options.display.width = 100
import matplotlib.pyplot as plt
import matplotlib
%matplotlib inline

import MDAnalysis as mda
from MDAnalysis import transformations
import nglview as nv

import os,re
%run -i _functions.py
```

* We use the system generated in [`example_water_between_two_walls`](../example_water_between_two_walls/water_between_walls.ipynb) with 9000 water molecules between two walls.
* To control the pressure in a mechanical way, we fix the bottom wall (or we can allow these atoms to shake around their inital position). Meanwhile, the top wall can move along the vertial direction (say z-direction) as a rigid body with a force from the external pressure imposed on it.

* `data.lmp` is generated in [`example_water_between_two_walls`](../example_water_between_two_walls/water_between_walls.ipynb) with 9000 water molecules
* `input.lmp` is the LAMMPS input script for run the simulation, please note that:
    * We extend the box in z direction at first so the top wall has space to flucuate to maintain the pressure. This is required when the long-range electrostatic force is calculated for a charged system because the periodic boudnary should be applied in all directions in this case. Otherwise, a non-periodic `s` can be used in LAMMPS and we need not to change the box size.
    * The external force from the pressure is imposed on the top wall via commands under `# control pressure`
    * We also compute and output the local stress tensor in the script.

```python
# map type and element

directory = './'
filename = directory+'data.lmp'
elements = fetchList(filename, 'Masses', 0,4, skipBlankSplit=True)
digitspattern = r'#'
types = []
chemical_symbols = []
for element in elements:
    types.append(int(element[0]))
    txt = re.sub(digitspattern, '', element[-1])
    chemical_symbols.append(atomic_symbol(txt))
types = np.array(types)

print('types:            ',types)
print('chemical_symbols: ',chemical_symbols)
```

## Analyse results

```python
# visualize trajectroy

h2o_num=9000 
# full: atom-ID molecule-ID atom-type q x y z
u = mda.Universe('result_atoms.data', atom_style='id resid type charge x y z')

## elements
elements = []
for i in range(len(u.atoms)):
    elements.append(chemical_symbols[np.where(types==int(u.atoms[i].type))[0][0]])
u.add_TopologyAttr('element',values=elements)

## resnames
resnames = ['up']+['low']+['h2o']*h2o_num
u.add_TopologyAttr('resnames',values=resnames)

u.load_new("result_atoms.lammpstrj", format="LAMMPSDUMP",timeunit="fs",dt=10000)
workflow = [transformations.unwrap(u.atoms)]
u.trajectory.add_transformations(*workflow)

print('dimensions:',u.dimensions)
print('frames',u.trajectory.n_frames)


view = nv.show_mdanalysis(u)
view.clear_representations()
view.add_representation(selection='h2o',repr_type='licorice',radius='.7',opacity=1)
view.add_representation(selection='up',repr_type='ball+stick',radius='1',opacity=1)
view.add_representation(selection='low',repr_type='ball+stick',radius='1',opacity=1)
view.add_unitcell()
view.camera = 'orthographic'
view.control.spin([1,0,0],-np.pi/2.)
view
```

```python
# render many frames and download images
from time import sleep
for frame in range(0, 10):
    view.frame = frame
    sleep(0.5)
    view.download_image(filename='image%04d.png'%frame,factor=8,antialias=True,trim=False,transparent=True)
    sleep(2.0)  
# then I make a gif with the rendered images
```

```python
# density of fluid and local pressure

names = ['id','z','n','sxx','syy','szz','den']
filename = 'result_profile_stress.log'
stress = pd.read_csv(filename,header=None,names=names,sep='\s+',skiprows=4)


fig = plt.figure(figsize=(8,2.2), dpi=140);

ax = fig.add_subplot(121);
x = stress['z']/10 #A2nm
pxx = stress['den']/3*1e3
ax.plot(x,pxx,color='tab:blue',alpha=1);

ax.plot((-10,10),(33.3,33.3),'k--',lw=1)

ax.set_xlabel('z');
ax.set_ylabel('density [nm$^{-3}$]');
ax.set_xlim(-4,2);
ax.set_ylim(0,50);
ax.tick_params(direction='in')


ax = fig.add_subplot(122);
    
#stress
x = stress['z']/10 #A2nm
pxx = -(stress['sxx']+stress['syy']+stress['szz'])/3*stress['den']
ax.plot(x,pxx,color='tab:blue',alpha=1);
 
ax.set_xlabel('z');
ax.set_ylabel('pressure [atm]');
ax.set_xlim(-4,2);
ax.set_ylim(-1e3,1e3);
ax.tick_params(direction='in')

plt.tight_layout()
plt.show()
```

* The bulk density of fluid is $\sim 33$ nm$^{-3}$, closing the bulk density of SPC/E water at $300$ K and $1$ atm, suggesting the pressure control works well.
* The local pressure profile shows significantly osillations, but we can see the pressure of bulk fluid fluctuates around $\sim 1$ atm.
